```python
""""punto 3======================================================="""

inventario = [
    {"nombre": "Pan",   "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café",  "precio": 8000, "stock": 10},
]


print("Primer producto:", inventario[0]["nombre"])


inventario.append({"nombre": "Arroz", "precio": 2500, "stock": 20})


for p in inventario:
    if p["nombre"] == "Leche":
        p["stock"] -= 2


total = 0
for p in inventario:
    total += p["precio"] * p["stock"]
print(f"Valor total del inventario: ${total}")


print("Stock bajo (< 10):")
for p in inventario:
    if p["stock"] < 10:
        print(" -", p["nombre"])


for p in inventario:
    p["precio"] = round(p["precio"] * 1.10, 2)


buscar = input("Ingrese nombre del producto a buscar: ")
encontrado = None
for p in inventario:
    if p["nombre"].lower() == buscar.lower():
        encontrado = p
        break

if encontrado:
    print("Producto encontrado:", encontrado)
else:
    print("Producto no encontrado.")


inventario = [p for p in inventario if p["stock"] != 0]


inventario.sort(key=lambda p: p["precio"])
print("Inventario ordenado por precio:", inventario)
```

```python
""""punto 4======================================================="""
# ── Punto 4: tuplas ──────────────────────────────────────────────────────────
import math

# Tupla: coordenadas (x, y)
punto = (3, 7)
print("Punto:", punto)
# Rectángulo
ancho = 5
alto  = 3
rectangulo = (ancho, alto)

area      = rectangulo[0] * rectangulo[1]
perimetro = 2 * (rectangulo[0] + rectangulo[1])
print(f"Área: {area}  |  Perímetro: {perimetro}")


try:
    rectangulo[0] = 10
except TypeError as e:
    print(f"Error al modificar la tupla: {e}")
    print("Las tuplas son inmutables: una vez creadas, sus valores no cambian.")
""""parte hecha con claude pq no entendi lo que habia que hacer"""


posiciones = [(0, 0), (1, 2), (3, 4)]
distancia_total = 0
for i in range(1, len(posiciones)):
    x1, y1 = posiciones[i - 1]
    x2, y2 = posiciones[i]
    distancia_total += math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print(f"Distancia total recorrida: {distancia_total}")
```

```python
""""punto 5====================================================="""
biblioteca = {
    "libros": [
        {"titulo": "1984",         "autor": "George Orwell",             "prestado": False},
        {"titulo": "El Principito","autor": "Antoine de Saint-Exupéry",  "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

prestados   = sum(1 for libro in biblioteca["libros"] if libro["prestado"])
disponibles = len(biblioteca["libros"]) - prestados
print(f"Prestados: {prestados}  |  Disponibles: {disponibles}")

# Agregar nuevo libro
biblioteca["libros"].append({"titulo": "Cien años de soledad","autor": "Gabriel García Márquez","prestado": False})


for libro in biblioteca["libros"]:
    if libro["titulo"] == "1984":
        libro["prestado"] = True


print(biblioteca)
```
