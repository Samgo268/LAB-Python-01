```python
"""punto 3======================================================="""
contador = 0
while contador < 5:
    print("Contador:", contador)
    contador += 1


contador = 5
while contador > 0:
    print("Contador:", contador)
    contador -= 1


while True:
    num = float(input("Número (0 para salir): "))
    if num == 0:
        break

# El estado del ciclo es la variable `contador` (o `num` y `acumulado`).
# Cambia en cada iteración y determina cuándo termina el while.
```

```python
"""punto 4======================================================="""
suma_impares = 0
suma_pares = 0
for i in range(1, 11):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

print("Suma de los pares:", suma_pares)
print("Suma de los impares:", suma_impares)

# Estado final como diccionario
estado_final = {"suma_pares": suma_pares, "suma_impares": suma_impares}
print("Estado final:", estado_final)

# En cada vuelta del bucle, i cambia y se acumula en la variable correcta.
# El estado evoluciona iteración a iteración hasta completar range(1,11).
```

```python
"""punto 5======================================================="""
productos = []
while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")
    if nombre == "fin":
        break
    precio   = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))

    
    if precio < 0 or cantidad < 0:
        print("Precio o cantidad inválidos, producto omitido.")
        continue

    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})


total = 0
for p in productos:
    total += p["precio"] * p["cantidad"]

print("Total del inventario:", total)


if productos:
    mas_caro = max(productos, key=lambda p: p["precio"])
    print("Producto más caro:", mas_caro)

print("Inventario final:", productos)
```
